```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
import datetime
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('subdivisions/temporary_database.csv', index_col=0)
df['year'] = df['year'].apply(lambda date: datetime.datetime.strptime(date,"%Y-%m-%d"))
df
```

```python
prod_names = df['product'].value_counts().index.to_list()
prod_list_by_name = {name: df[df['product'] == name].sort_values('destinated_area', ascending=False) for name in prod_names}
for name, prod in prod_list_by_name.items():
    y = prod['destinated_area']
    x = prod['year']
    fig = plt.figure(figsize=(6,6))
    ax = fig.add_subplot(111)
    ax.scatter(x, y, alpha=0.4, lw=0)
    ax.set_xlabel('Time')
    ax.set_ylabel('Area')
    ax.set_title(name)
    ax.grid(True)
```
